```python
# Python code to demonstrate the working of 
# mean()

# importing statistics to handle statistical
# operations 
import statistics

# initializing list 
li = [1, 2, 3, 3, 2, 2, 2, 1, 2, 3, 4, 5, 7, 9, 11] 

# using mean() to calculate average of list
# elements 
print ("The average of list values is : ",end="") 
print(statistics.mean(li))
print ("The median of list values is : ",end="\n") 
print(statistics.median(li))
print ("The mode of list values is : ",end="") 
print(statistics.mode(li))
print ("The variance of list values is : ",end="") 
print(statistics.variance(li))
print ("The stdev of list values is : ",end="") 
print(statistics.stdev(li))
```

```python
from scipy.stats import binom, norm, poisson
import numpy as np
import matplotlib.pyplot as plt
data_norm = np.random.normal(0,1,1000)
plt.hist(data_norm, bins = 30, density = True)
bino = binom.rvs(n=10, p = 0.3, size = 1000)
plt.hist(bino, bins = 10, density = True, alpha = 0.5)
poiss = poisson.rvs(mu=3, size = 1000)
plt.hist(poiss, bins = 15, density = True, alpha = 0.5)
plt.legend(['Normal', 'Binomial', 'Poission'])
plt.show()
```

```python
data1 = 20 * np.random.randn(1000) + 100
data2 = data1 + (10 * np.random.randn(1000) + 50)
# summarize
print('data1: mean=%.3f stdv=%.3f' % (np.mean(data1), np.std(data1)))
print('data2: mean=%.3f stdv=%.3f' % (np.mean(data2), np.std(data2)))
# plot
# calculate covariance matrix
covariance = np.cov(data1, data2)
print(covariance)
corr, _ = scipy.stats.pearsonr(data1, data2)
print('Pearsons correlation: %.3f' % corr)
# calculate spearman's correlation
corr, _ = scipy.stats.spearmanr(data1, data2)
print('Spearmans correlation: %.3f' % corr)
plt.scatter(data1, data2)
plt.show()
```
